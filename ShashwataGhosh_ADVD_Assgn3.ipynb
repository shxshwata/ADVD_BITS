```python
import numpy as np
import matplotlib.pyplot as plt

Vt = 0.2
beta = 1
lambda_val = 0.7
VGS_values = [0.2, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0]
VDS = np.linspace(0, 1, 200)
plt.figure(figsize=(8,6))
for VGS in VGS_values:
    ID = []
    for vds in VDS:
        if VGS <= Vt:
            id_val = 0
        elif vds < (VGS - Vt):
            id_val = beta * ((VGS - Vt) * vds - (vds**2)/2) * (1 + lambda_val*vds)
        else:
            id_val = 0.5 * beta * (VGS - Vt)**2 * (1 + lambda_val*vds)
        ID.append(id_val)
    plt.plot(VDS, ID, label=f"VGS={VGS:.1f} V")
plt.xlabel(r"$V_{DS}$ [V]")
plt.ylabel(r"$I_D$ [mA/m]")
plt.title("MOSFET $I_D - V_{DS}$ with Channel Length Modulation (=0.7)")
plt.legend()
plt.grid(True)
plt.show()
```
