```python
import numpy as np
import matplotlib.pyplot as plt
Vt = 0.2
beta = 1
lam = 0
VGS_values = [0.2, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0]
VDS = np.linspace(0, 1, 100)
plt.figure(figsize=(8, 5))
for VGS in VGS_values:
    ID = []
    for vds in VDS:
        if VGS <= Vt:
            id_val = 0
        elif vds < (VGS - Vt):  # Triode region
            id_val = beta * ((VGS - Vt) * vds - 0.5 * vds**2)
        else:  # Saturation region
            id_val = 0.5 * beta * (VGS - Vt)**2 * (1 + lam * vds)
        ID.append(id_val)
    plt.plot(VDS, ID, label=f"$V_{{GS}}$={VGS:.1f} V")
plt.xlabel(r"$V_{DS}$ [V]")
plt.ylabel(r"$I_D$ [mA/$\mu$m]")
plt.legend()
plt.grid(True)
plt.show()
```
