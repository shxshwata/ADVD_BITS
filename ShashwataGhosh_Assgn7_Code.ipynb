```python
#Shashwata Ghosh-2022B5A31036H
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
string = r"C:\Users\MacBook\Downloads\sem 7\ADVD\ShashwataGhosh_ADVDAsgn7.csv"
datas = pd.read_csv(string)
datas.head()
```

```python
datas['Ratio'] = datas['Wp (µm)'] / datas['Wn (µm)']
X = datas[['Vmid(V)']]
y = datas['Ratio']

print("Data with new 'Ratio' column:")
print(datas.head())
```

```python
poly_degree = 3
poly = PolynomialFeatures(degree=poly_degree, include_bias=False)
X_poly = poly.fit_transform(X)

model = LinearRegression()
model.fit(X_poly, y)

print(f"Successfully trained a polynomial model of degree {poly_degree}.")
```

```python

VDD = 1.8
target_vmid = VDD / 2.0

target_vmid_poly = poly.transform([[target_vmid]])

predicted_ratio_array = model.predict(target_vmid_poly)
model_predicted_ratio = predicted_ratio_array[0]

print("\n--- AI Model Prediction (assignment 07) ---")
print(f"Target Vmid (Vdd/2) is: {target_vmid:.4f} V")
print(f"The AI model predicts a Wp/Wn Ratio of: {model_predicted_ratio:.4f}")
print("--------------------------------------")


manual_estimate_ratio = 3.288
difference = model_predicted_ratio - manual_estimate_ratio
percent_error = (abs(difference) / manual_estimate_ratio) * 100

print("\n--- Comparison with Assignment 5 ---")
print(f"Manual Estimate (assignment 05):   {manual_estimate_ratio:.4f}")
print(f"AI Model Prediction (assignment 07): {model_predicted_ratio:.4f}")
print(f"Difference (AI - Manual):     {difference:.4f}")
print(f"Percent Error:                {percent_error:.2f}%")
print("------------------------------------")
```

```python

X_plot = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)

X_plot_poly = poly.transform(X_plot)

y_plot = model.predict(X_plot_poly)

plt.scatter(datas['Vmid(V)'], datas['Ratio'], label='Actual Data')

plt.plot(X_plot, y_plot, color='red', label=f'Polynomial Fit (degree {poly_degree})')

predicted_ratio = model.predict(target_vmid_poly)

plt.plot(target_vmid, predicted_ratio, 'r*', markersize=15, label=f'Predicted Point (Vmid={target_vmid:.2f}V)')

plt.xlabel('Vmid (V)')
plt.ylabel('Wp/Wn Ratio')
plt.title('Polynomial Regression: Vmid vs. Wp/Wn Ratio')
plt.legend()
plt.grid(True)
plt.show()
```
